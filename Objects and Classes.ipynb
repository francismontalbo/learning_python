# Constructors in Python

```python
class ComplexNumber:
    def __init__(self, r=0, i=0):
        self.real = r
        self.imag = i

    def get_data(self):
        print(f'{self.real}+{self.imag}j')
```

```python
# Create a new ComplexNumber object
num1 = ComplexNumber(2, 3)
```

```python
# Call get_data() method
# Output: 2+3j
num1.get_data()
```

```python
# Create another ComplexNumber object
# and create a new attribute 'attr'
num2 = ComplexNumber(5)
num2.attr = 10
```

```python
# Output: (5, 0, 10)
print((num2.real, num2.imag, num2.attr))
```

```python
# but c1 object doesn't have attribute 'attr'
# AttributeError: 'ComplexNumber' object has no attribute 'attr'
# print(num1.attr)
```

# An interesting thing to note in the above step is that attributes of an object can be created on the fly. We created a new attribute attr for object num2 and read it as well. But this does not create that attribute for object num1.

# LET'S CREATE, FUEL, AND DRIVE A TESLA!!

```python
class Vehicle:
    def __init__(self, brand, model, year):
        self.brand = brand      
        self.model = model
        self.year = year
        self.car_type = 'gas'
        self.max_level = 15
        self.charge_level = 0
        
        if self.brand == 'tesla' or self.brand == 'Tesla':
            self.car_type = 'electric'
        else:
            self.car_type = 'gas'
              
    def get_info(self):
        if self.car_type == 'gas':
            print('The car brand is:', self.brand.title())
            print('The model is:', self.model.title())
            print('The model year is:', self.year)
            print('Max tank size is:', self.max_level)
            print('The car type is:', self.car_type)
            
        elif self.car_type == 'electric':
            print('The car brand is:', self.brand.title())
            print('The model is:', self.model.title())
            print('The model year is:', self.year)
            print('Max battery size is:', self.max_level)
            print('The car type is:', self.car_type)

    def fuel_up(self):
        if self.car_type == 'gas':
            self.charge_level = self.max_level
            print('Gas tank is now full.')
        
        elif self.car_type == 'electric':
            self.charge_level = self.max_level
            print(f'Battery charge is {self.charge_level} / {self.max_level}  >> FULL!')

    def drive(self):
        if self.car_type == 'gas':
            if self.charge_level <= 0:
                print(f"Your {self.model.title()} {self.brand.title()} has {self.charge_level} / {self.max_level} fuel!\nyou can't drive!\nGet gas first!")
                engineok.stop()
                enginebad.stop()
                enginebad.play()
            elif self.charge_level > 0:
                print('*'*100)
                print(f'Your {self.model.title()} {self.brand.title()} is now driving.')
                print('*'*100)
                engineok.stop()
                enginebad.stop()
                engineok.play()
                gas_consume = 5
                new_level = (self.charge_level - gas_consume) 
                self.charge_level = new_level
                print(f'Your {self.model.title()} {self.brand.title()} used about {abs(gas_consume)}x gas.')
                print(f'Your current fuel level is at {new_level} / {self.max_level}.')
                if self.charge_level <= 5:
                    print('You need to get some gas mate!')
                    print('You only have one drive left!')
                elif self.charge_level > 5:
                    print('You have enough gas! Keep rolling!')
                    
        elif self.car_type == 'electric':
            if self.charge_level <= 0:
                print(f"Your {self.model.title()} {self.brand.title()} has {self.charge_level} / {self.max_level} charge!\nyou can't drive!\nGet recharged first!")
                enginebad.stop()
                engineok.stop()
                enginebad.play()
            elif self.charge_level > 0:
                print('*'*100)
                print(f'Your {self.model.title()} {self.brand.title()} is now driving.')
                print('*'*100)
                engineok.stop
                enginebad.stop()
                engineok.play()
                battery_consume = 5
                new_level = (self.charge_level - battery_consume) 
                self.charge_level = new_level
                print(f'Your {self.model.title()} {self.brand.title()} used about {abs(battery_consume)}x battery charge.')
                print(f'Your current battery level is at {new_level} / {self.max_level}.')
            if self.charge_level <= 5:
                print('You need to get some charge mate!')
                print('You only have one drive left!')
            elif self.charge_level > 5:
                print('You have enough battery! Keep rolling!')
                
    def update_fuel_level(self, new_level):
        if self.car_type == 'gas':
            if new_level <= self.max_level:
                self.charge_level = new_level
            else:
                print('Exceeded tank capacity')
        elif self.car_type == 'electric':
            if new_level <= self.max_level:
                self.charge_level = new_level
            else:
                print('Exceeded battery capacity')
            
    def pump_up(self):
        engineok.stop
        enginebad.stop()
        if self.car_type == 'gas':
            gas_amount = int(input('How much gas you need? '))
            if (self.charge_level + gas_amount <= self.max_level):
                self.charge_level += gas_amount 
                print(f'Added fuel! Fuel level is now: {self.charge_level} / {self.max_level}.')
            else:
                print(f'The tank wont hold that much your max tank is only up to {self.max_level}.')
                
        elif self.car_type == 'electric':
            battery_charge = int(input('How much battery you need to charge? '))
            if (self.charge_level + battery_charge <= self.max_level):
                self.charge_level += battery_charge 
                print(f'Added charge! Battery level is now: {self.charge_level} / {self.max_level}.')
            else:
                print(f'The battery wont hold that much your max charge is only up to {self.max_level}.')
                
    def check_level(self):
        engineok.stop
        enginebad.stop()
        if self.car_type == 'gas':
            print(f'Your current gas level is {self.charge_level} / {self.max_level}')
            if self.charge_level <= 5:
                print(f'Please get some gas!, you only have {self.charge_level} / {self.max_level} left!')
            elif self.charge_level > 5:
                print(f'You can still go for another drive, your fuel level is still at {self.charge_level}')
        elif self.car_type == 'electric':
            print(f'Your current battery level is {self.charge_level} / {self.max_level}')
            if self.charge_level <= 5:
                print(f'Please charge your battery, you only have {self.charge_level} / {self.max_level} left!')
            elif self.charge_level > 5:
                print(f'You can still go for another drive, your fuel level is still at {self.charge_level} / {self.max_level}')
```
